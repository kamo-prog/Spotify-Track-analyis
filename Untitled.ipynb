```python
data.dropna(inplace = True)
```

```python
data.drop_duplicates(inplace = True)
```

```python
top_10 = data.sort_values("popularity", ascending = False).head(10)
sns.barplot(x = "popularity" , y = "track_name",hue = "artists", data = top_10, palette = "viridis")
plt.title("Top 10 Most Popular songs on sportify")
plt.show()
```

```python
plt.figure(figsize  = (10,6))
sns.scatterplot(x = "danceability", y = "energy",hue = "popularity", data = data, palette = "coolwarm")
plt.title("Dance vs. Energy")
plt.show()
```

```python
plt.figure(figsize = (12,8))
corri = data[["danceability", "energy", "loudness", "valence", "tempo", "popularity"]].corr()
sns.heatmap(corri, cmap = "coolwarm")
plt.title("Correlation between audio Features")
plt.show()
```

```python
##############Convert 'duration_ms' to minutes###############
data["duration_min"] = data ["duration_ms"]/60000
#############################################################
plt.figure(figsize=(10, 6)) 
sns.histplot(data['duration_min'], bins=30, kde=True)
plt.title('Distribution of Song Duration (Minutes)')
plt.xlabel('Duration (minutes)')
plt.show()
```

```python
#Duration VS Popularity 
plt.figure(figsize=(10, 6))
sns.scatterplot(x='duration_min', y='popularity', data=data, alpha=0.5)
plt.title('Song Duration vs. Popularity')
plt.xlabel('Duration (minutes)')
plt.show()
```
